```python
from PIL import Image

# 打开图像文件
image = Image.open("吃東C.png")

# 定义黄色的RGB颜色值
yellow_rgb = (255, 242, 0)  # 在RGB中，黄色通常是(255, 255, 0)

# 将图像中的黄色区域替换为灰色
def replace_color(img, target_color, replacement_color):
    img = img.convert("RGBA")
    data = img.getdata()
    new_image_data = []
    for item in data:
        # 如果颜色与目标颜色匹配，则替换为灰色，否则保留原始颜色
        if item[:3] == target_color:
            new_image_data.append(replacement_color + (item[3],))
        else:
            new_image_data.append(item)
    img.putdata(new_image_data)
    return img

gray_image = replace_color(image, yellow_rgb, (128, 128, 128))

# 显示修改后的图像
display(gray_image)
#在上述示例中，我们首先打开图像，然后定义了黄色的 RGB 颜色值 (255, 255, 0)。接下来，我们使用 replace_color 函数来替换图像中的黄色区域为灰色。最后，我们显示了修改后的图像。

#请确保将 "your_image.jpg" 替换为您实际的图像文件路径，并根据需要调整颜色值。这个示例中的 replace_color 函数可以用于替换图像中的特定颜色，您可以根据需要进行自定义。





```
